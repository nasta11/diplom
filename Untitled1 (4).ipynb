```python
import numpy as np
import pandas as pd

# Загрузка данных из CSV
dataset = pd.read_csv('C:\\Users\\nasta\\OneDrive\\Desktop\\dataset.csv')

# Проверка заголовков столбцов
print(f'Заголовки столбцов: {dataset.columns}')

# Названия столбцов
feature_column = 'feature'
target_column = 'target'

# Масштабирование данных
def min_max_scale(data):
    min_val = data.min()
    max_val = data.max()
    return (data - min_val) / (max_val - min_val)

scaled_feature = min_max_scale(dataset[feature_column])
scaled_target = min_max_scale(dataset[target_column])

# Определение длины данных для обучения
training_data_len = int(np.ceil(len(dataset) * 0.8))
print(f'Training data length: {training_data_len}')

# Создание тренировочного набора данных
train_feature = scaled_feature[:training_data_len]
train_target = scaled_target[:training_data_len]

# Разделение данных на x_train и y_train
x_train = []
y_train = []

# Шаг для формирования данных
step_length = 10

# Проверка содержимого train_feature перед циклом
if len(train_feature) <= step_length:
    raise ValueError("Недостаточно данных для формирования обучающей выборки")

for i in range(step_length, len(train_feature)):
    x_train.append(train_feature[i-step_length:i])
    y_train.append(train_target[i])

x_train, y_train = np.array(x_train), np.array(y_train)

# Добавление константы для регрессионной модели
X_train = np.hstack([np.ones((x_train.shape[0], 1)), x_train])  # Добавляем столбец единиц

# Параметр регуляризации для Ridge-регрессии
alpha = 1.0

# Решение линейной регрессии с регуляризацией (Ridge-регрессия)
# Формула: beta = (X^T * X + alpha * I)^(-1) * X^T * y
I = np.eye(X_train.shape[1])
X_transpose = X_train.T
coefficients = np.linalg.inv(X_transpose @ X_train + alpha * I) @ X_transpose @ y_train

# Подготовка тестовых данных
test_feature = scaled_feature[training_data_len - step_length:]

# Подготовка x_test
x_test = []
y_test = scaled_target[training_data_len:]

for i in range(step_length, len(test_feature)):
    x_test.append(test_feature[i-step_length:i])

x_test = np.array(x_test)
X_test = np.hstack([np.ones((x_test.shape[0], 1)), x_test])  # Добавляем столбец единиц

# Прогнозирование
predictions = X_test @ coefficients

# Преобразование предсказаний и y_test в правильный формат
predictions = predictions.reshape(-1, 1)
y_test = np.array(y_test[:len(predictions)]).reshape(-1, 1)  # Преобразуем y_test в numpy массив и затем изменяем форму

# Обратное масштабирование предсказаний и y_test
def inverse_min_max_scale(scaled_data, original_data):
    min_val = original_data.min()
    max_val = original_data.max()
    return scaled_data * (max_val - min_val) + min_val

try:
    predictions = inverse_min_max_scale(predictions, dataset[feature_column])
    y_test = inverse_min_max_scale(y_test, dataset[target_column])
except ValueError as e:
    print(f'Ошибка при обратном масштабировании: {e}')
    print(f'Размеры predictions: {predictions.shape}, y_test: {y_test.shape}')

# Убедитесь, что размеры совпадают
if predictions.shape != y_test.shape:
    raise ValueError(f'Размеры predictions и y_test не совпадают: {predictions.shape} vs {y_test.shape}')

# Вычисление RMSE
rmse = np.sqrt(np.mean((predictions - y_test) ** 2))
print(f'Root Mean Squared Error: {rmse}')

# Вывод результатов
print(predictions)


```

```python
# Define the function to return the SMAPE value
def calculate_smape(actual, predicted) -> float:

    # Convert actual and predicted to numpy
    # array data type if not already
    if not all([isinstance(actual, np.ndarray),
                isinstance(predicted, np.ndarray)]):
        actual, predicted = np.array(actual),
        np.array(predicted)

    return round(
        np.mean(
            np.abs(predicted - actual) /
            ((np.abs(predicted) + np.abs(actual))/2)
        )*100, 2
    )
```

Вы можете научить свою модель прогнозировать будущую последовательность (например, следующие 30 дней) вместо прогнозирования следующего значения (на следующий день), как это происходит в настоящее время.

Для этого вам необходимо определить выходные данные как y[t: t + H](вместо y[t]того, как в текущем коде), где y– временной ряд и H– продолжительность периода прогноза (т. е. количество дней вперед, на которое вы хотите спрогнозировать). Также необходимо установить количество выходов последнего слоя равным H(а не равным 1как в текущем коде).

Вы по-прежнему можете определить входные данные как y[t - T: t]продолжительность Tпериода ретроспективного анализа (или количество временных шагов), и поэтому форма входных данных модели по-прежнему равна (T, 1). Период ретроспективного анализа Tобычно длиннее периода прогноза H(т. е. T > H) и часто устанавливается равным кратному H(т. е. T = m * Hгде m > 1– целое число).
